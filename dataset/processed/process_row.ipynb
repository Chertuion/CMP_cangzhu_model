```python
import pandas as pd
import re


file_path = '/data/home/wxl22/changzu_ans/dataset/processed/new/98地区重复GC76成分.xlsx'
df = pd.read_excel(file_path)


df['SamplePrefix'] = df['Sample'].str.extract(r'^(.*?)(?:-\d)$')

def format_sample_id(sample):

    match = re.match(r'^([A-Za-z]+)(\d+)$', sample)
    if match:
        prefix, number = match.groups()
        if len(number) == 1:
            return f"{prefix}0{number}"
        return f"{prefix}{number}"
    return sample  


df['SamplePrefix'] = df['SamplePrefix'].apply(format_sample_id)


numeric_columns = df.select_dtypes(include='number').columns
result = df.groupby('SamplePrefix')[numeric_columns].mean().reset_index()


result.rename(columns={'SamplePrefix': 'Sample'}, inplace=True)


output_file = '/data/home/wxl22/changzu_ans/dataset/processed/97AC.xlsx'
result.to_excel(output_file, index=False)

```
